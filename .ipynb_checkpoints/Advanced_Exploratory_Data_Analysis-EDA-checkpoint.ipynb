#### Below code will shows the list of features which has the missing values

```python
features_with_na = [features for features in df.columns if df[features].isnull().sum()>1]
features_with_na
```

#### Below code will show the percentage of nan values present in each feature

```python
for feature in features_with_na:
    print(feature,np.round(df[feature].isnull().mean(),4),"%missing values")
```

#### Finding the relationship between missing values and Sales Price

```python
for feature in features_with_na:
    data = df.copy()
    
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar(color = 'b')
    plt.title(feature)
    plt.show()
```

 The relation between the missing values and the dependent variable is clearly visible.

#### To find the Numerical Variables columns in the Data Set

```python
numerical_feature = [feature for feature in df.columns if df[feature].dtypes != 'O']
```

```python
print("No of Numerical Feature is : {}".format(len(numerical_feature)))
```

From the Dataset we have 4 year variables. We have to extract information from the datetime variables like no of years or no of days.

```python
year_feature = [feature for feature in numerical_feature if 'Year' in feature or 'Yr' in feature or 'year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature,df[feature].unique())
```

### Analyzing the Datetime Variables

Checking whether there is a relation between year of the house is sold and the sales price

```python
df.groupby('YrSold')['SalePrice'].median().plot(color = ['blue'])
plt.title('House Price vs Year Sold')
plt.xlabel('Year Sold')
plt.ylabel('Sale Price')
plt.grid()
plt.show()
```

#### Comparing the difference between all years feature with SalePrice

```python
year_feature
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=df.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.show()
```

### Getting the Discrete variables from the Data Set

```python
discrete_feature = [feature for feature in numerical_feature if len(df[feature].unique())<25 and feature not in year_feature + ['Id']]
print("No of Discrete feature are : {}".format(len(discrete_feature)))
```

```python
df[discrete_feature].head()
```

### Find the realtionship between discrete_feature and Sale Price

```python
discrete_feature
```

```python
for feature in discrete_feature:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['red','orange','blue','green','yellow'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

### Getting the Continuous variable from the Data Set

```python
continuous_feature = [feature for feature in numerical_feature if feature not in discrete_feature and feature not in year_feature + ['Id'] ]
print("No of Continuous Features are {}".format(len(continuous_feature)))
```

```python
for feature in continuous_feature:
    data = df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.title(feature)
    plt.ylabel('count')
    plt.show()
```

### will be using logarithmic transformation

```python
for feature in continuous_feature:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.title(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

### Finding the Outliers using box plots

```python
for feature in continuous_feature:
    data = df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

### Getting the Categorical Variables from the Data Set

```python
categorical_feature = [feature for feature in df.columns if df[feature].dtypes == 'O']
```

```python
for feature in categorical_feature:
    print('The feature is {} and number of categories are {}'.format(feature,len(df[feature].unique())))
```

### Relationship between categorical variable and dependent feature SalesPrice

```python
for feature in categorical_feature:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['red','orange','blue','green','yellow'])
    plt.xlabel(feature)
    plt.title(feature)
    plt.ylabel('SalePrice')
    plt.show()
```
